# Setting this up

We are going to make all our requests to mongod rest api using [requests library](http://docs.python-requests.org/en/latest/) 

```python
import requests
import simplejson as json

def pprint(s):
    print(json.dumps(s, sort_keys=True, indent=4, separators=(',', ':')))

#call mongod stats using the rest api
uri = "http://localhost:28017/"
r = requests.get(uri+'serverStatus')

assert 200, r.status_code
#collect the api response
response = r.json()

pprint(response)
```
