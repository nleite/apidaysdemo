#Testing the Ops Manager API
Ops Manager comes with a strong and extensive API that is fully [documented](https://docs.mms.mongodb.com/reference/api/hosts/) 
Feel free to dig around to check the info available

##Start connecting to the API 

```python
import requests
from requests.auth import HTTPDigestAuth
import simplejson as json
#pretty print method for json
def pprint(s):
    print(json.dumps(s, sort_keys=True, indent=4, separators=(',', ':')))
```

##Secure API
Ops Manager API enforces an http digest authentication based on API keys that the user needs to create before enabling the requests

```python
username = 'YOUR_USER'
pass_key = 'YOUR_API_KEY'
host = 'YOUR_HOST'
digest_auth = HTTPDigestAuth(username, pass_key)
```

##Root End Point
We can start by connecting to the [root endpoint](https://docs.mms.mongodb.com/reference/api/root/)

```python


api_r = requests.get( host+"/api/public/v1.0" , auth=digest_auth )

pprint(api_r.json())
```

##Browsable API
we can use the `links` attribute of the api response to *nagivate* through the exposed resources

```python
further_urls = []

for l in api_r.json()['links']:
    if l['rel'] != 'self':
        further_urls.append(l['href'])

further_urls

for u in further_urls:
    u_r = requests.get(u, auth=digest_auth)
    pprint(u_r.json())
```

##Deployment State
We can verify several actions that are occuring within a deployment, like in which state is it now

```python
#find my groups
uri_usergroups = '%s/api/public/v1.0/groups' % (host)

group_id =  requests.get(uri_usergroups, auth=digest_auth).json()['results'][0]['id'] #"554a44190cf23b49891dad16"
uri_groupstate = '%s/api/public/v1.0/groups/%s/automationStatus' % (host, group_id)
deploymentstate_r = requests.get(uri_groupstate, auth=digest_auth)
pprint(deploymentstate_r.json())
```

###Pretty Replies
Replies can be pretty out of the box `pretty=true`

```python
user_id = "554a46cc0cf23b49891dad5d"
uri_user ='%s/api/public/v1.0/users/%s?pretty=true' % (host, user_id)
user_r = requests.get(uri_user, auth=digest_auth)

print user_r.text
```

##Perform cluster changes
Obvilously that a system that only allows to read is nice, but we want to read and write.
I want to add a new member to my replica set 

```python
uri_groupconfig = '%s/api/public/v1.0/groups/%s/automationConfig' % (host, group_id)

config_r = requests.get(uri_groupconfig, auth=digest_auth)

config_json = config_r.json()

pprint(config_json)
```

Once we have the existing configuration we will be performing a `put` with the new replicaset member

```python
new_replica_member = {'_id': 3, 'hidden':True, 'host':'apidaysrepl_3'}
config_json['replicaSets'][0]['members'].append( new_replica_member )
```

And a new process where the replica member will be running

```python
import copy 
#need to copy the dictionary otherwise going to be changing the previous reference
if len(config_json['processes']) < 4:
    new_replica_process = copy.deepcopy(config_json['processes'][0])
    #set port 
    new_replica_process['args2_6']['net']['port'] = 27003
    new_replica_process['args2_6']['storage']['dbPath'] = '/Users/norberto/sandbox/apidays/replset/apidaysrepl_3'
    new_replica_process['args2_6']['systemLog']['path'] = '/Users/norberto/sandbox/apidays/replset/apidaysrepl_3/mongodb.log'
    new_replica_process['name'] = 'apidaysrepl_3'

    config_json['processes'].append(new_replica_process)
pprint(config_json)
```

```python
config_p = requests.put(uri_groupconfig, auth=digest_auth, data=json.dumps(config_json), headers={'content-type': 'application/json'} )
print config_p.text
```
